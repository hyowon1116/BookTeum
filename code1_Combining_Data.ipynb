```python
import os
import pandas as pd
from tqdm import tqdm

import warnings
warnings.filterwarnings('ignore')
```

```python
path = './data/rawdata/'

df = pd.DataFrame()
files = [file for file in os.listdir(path) if file.startswith('NL_BO_SPECIES')] # csv 파일명 리스트

files
```

```python
for file in tqdm(files):
    df = pd.concat([df, pd.read_csv(path + file)]) # 순서대로 데이터 병합
```

```python
df = df.reset_index(drop = True) # 인덱스 초기화
df
```

```python
df.info() # 데이터 정보 확인
```

```python
df_path = './data/Result1_Combined.csv'
df.to_csv(df_path, encoding = 'utf-8-sig', index = None) # csv로 저장
```
